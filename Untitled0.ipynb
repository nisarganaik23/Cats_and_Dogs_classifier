```python
import os

base_dir = '/content/datasets/cats_and_dogs_filtered'  # Path to the extracted dataset folder

# List contents of the base directory
print("Base directory contents:", os.listdir(base_dir))

# List contents of the 'train' and 'validation' directories
train_dir = os.path.join(base_dir, 'train')
val_dir = os.path.join(base_dir, 'validation')

print("Train directory contents:", os.listdir(train_dir))
print("Validation directory contents:", os.listdir(val_dir))
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Set image dimensions and batch size
IMG_HEIGHT = 150
IMG_WIDTH = 150
BATCH_SIZE = 32

# Initialize the ImageDataGenerator for rescaling the images
train_image_generator = ImageDataGenerator(rescale=1./255)  # Rescaling images for training
val_image_generator = ImageDataGenerator(rescale=1./255)  # Rescaling images for validation

# Create the training and validation data generators
train_data_gen = train_image_generator.flow_from_directory(
    train_dir,  # Path to train directory
        target_size=(IMG_HEIGHT, IMG_WIDTH),  # Resize the images
            batch_size=BATCH_SIZE,  # Set the batch size
                class_mode='binary'  # Use binary classification (cat or dog)
                )

val_data_gen = val_image_generator.flow_from_directory(
                    val_dir,  # Path to validation directory
                        target_size=(IMG_HEIGHT, IMG_WIDTH),  # Resize the images
                            batch_size=BATCH_SIZE,  # Set the batch size
                                class_mode='binary'  # Use binary classification (cat or dog)
                                )
```

```python
from tensorflow.keras import layers, models

# Build the model
model = models.Sequential([
    # First Convolutional Layer
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
            layers.MaxPooling2D(2, 2),

                # Second Convolutional Layer
                    layers.Conv2D(64, (3, 3), activation='relu'),
                        layers.MaxPooling2D(2, 2),

                            # Third Convolutional Layer
                                layers.Conv2D(128, (3, 3), activation='relu'),
                                    layers.MaxPooling2D(2, 2),

                                        # Flatten the output
                                            layers.Flatten(),

                                                # Fully connected layer
                                                    layers.Dense(512, activation='relu'),

                                                        # Output layer (binary classification: cat or dog)
                                                            layers.Dense(1, activation='sigmoid')
])

                                                            # Compile the model
model.compile(optimizer='adam',
                                                                          loss='binary_crossentropy',
                                                                                        metrics=['accuracy'])

                                                                                        # Summary of the model
model.summary()
```

```python
# Set the number of steps per epoch and validation steps
train_steps = train_data_gen.samples // train_data_gen.batch_size
val_steps = val_data_gen.samples // val_data_gen.batch_size

# Train the model
history = model.fit(
    train_data_gen,  # Training data generator
        steps_per_epoch=train_steps,  # Steps per epoch
            epochs=10,  # Number of epochs to train the model
                validation_data=val_data_gen,  # Validation data generator
                    validation_steps=val_steps  # Steps for validation
                    )
```

```python
# Evaluate the model on the validation data
val_loss, val_accuracy = model.evaluate(val_data_gen, steps=val_steps)

print(f"Validation Loss: {val_loss}")
print(f"Validation Accuracy: {val_accuracy}")
```
